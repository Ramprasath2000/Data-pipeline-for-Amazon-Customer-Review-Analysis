```python
import pandas as pd
df = pd.read_csv("/content/aws_data - Sheet1.csv")
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
# Check for empty strings, "NA", "NaN", etc.
hidden_nulls = df.apply(lambda x: x.astype(str).str.contains("NA|NaN|n/a|^$", na=False))
print("Potential hidden 'null' strings:")
print(hidden_nulls.sum())
```

```python
df.info()
```

```python
#converting to standard date time
df["review_date"] = pd.to_datetime(df["review_date"])

#if it is in object it will be usefull
# Extract date components
df['review_year'] = df['review_date'].dt.year
df['review_month'] = df['review_date'].dt.month
#cusotmer id
df["customer_id"] = df["customer_id"].astype(object)
#only 3 values exist so it is better to use category
df["marketplace"] = df["marketplace"].astype("category")
df["product_category"] = df["product_category"].astype("category")
df["product_parent"] = df["product_parent"].astype("int64")
# display(df["product_category"].unique())
```

```python
# Then keep only the date component
df["review_date"] = df["review_date"].dt.date
```

```python
#standardisation check
print("Unique Product Categories:")
print(df["product_category"].unique())

print("\nUnique Marketplaces:")
print(df["marketplace"].unique())
```

```python
#backing up the original to preserve readability
df['review_body_raw'] = df['review_body']
df['review_headline'] = df['review_headline']
```

```python
#standarzing to lower case
df['review_body'] = df['review_body'].str.lower()
df['review_headline'] = df['review_headline'].str.lower()
```

```python
#removing punctuations
import string

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['review_body'] = df['review_body'].apply(remove_punctuation)
df['review_headline'] = df['review_headline'].apply(remove_punctuation)
```

```python
#remove whitespaces
df['review_body'] = df['review_body'].str.strip().replace(r'\s+', ' ', regex=True)
df['review_headline'] = df['review_headline'].str.strip().replace(r'\s+', ' ', regex=True)
```
