```python
# Identify the products with the highest number of reviews.
query = """
SELECT
    product_id,
    product_title,
    COUNT(*) AS review_count,
    AVG(star_rating) AS avg_rating
FROM
    product_reviews
GROUP BY
    product_id, product_title
ORDER BY
    review_count DESC
LIMIT 10;
"""

top_products_df = pd.read_sql(query, engine)
display(top_products_df)
```

```python
# The Average Review Ratings Per Month for Each Product
# Task: Calculate the average star rating per month for each product.
# Outcome: Understand the trend in average ratings over time for different products.
from sqlalchemy import text
query = text("""
SELECT
    product_id,
    product_title,
    DATE_FORMAT(review_date, '%Y-%m') AS review_month,
    AVG(star_rating) AS avg_rating,
    COUNT(*) AS review_count
FROM
    product_reviews
GROUP BY
    product_id,
    product_title,
    DATE_FORMAT(review_date, '%Y-%m')
ORDER BY
    product_id,
    review_month;
    """)

avg_rating_per_month_df = pd.read_sql(query, engine)
display(avg_rating_per_month_df)
```

```python
# The Total Number of Votes Per Product Category
# Task: Aggregate the total votes per product category.
# Outcome: Identify the most popular product categories based on total votes.
from sqlalchemy import text
query = text("""
SELECT
    product_category,
    SUM(total_votes) AS votes_count
FROM
    product_reviews
GROUP BY
    product_category

ORDER BY
    votes_count DESC;
    """)

votes_per_category_df = pd.read_sql(query, engine)
display(votes_per_category_df)
```

```python
# Products with the Word "Awful" Occurring Most Frequently in Reviews
# Task: Identify products with the most mentions of the word "awful" in their reviews.
# Outcome: Understand customer sentiment towards products associated with the word "awful".
# Outcome: Identify the most popular product categories based on total votes.
from sqlalchemy import text
query = text("""
SELECT
    product_id,
    product_title,
    COUNT(*) AS awful_count
FROM
    product_reviews
WHERE
    review_body LIKE '%awf%'
GROUP BY
    product_id,
    product_title
ORDER BY
    awful_count DESC;
    """)

awful_products_df = pd.read_sql(query, engine)
display(awful_products_df)
```

```python
# Task: Identify products with the most mentions of the word "awesome".
# Outcome: Understand customer sentiment towards products associated with the word "awesome".
from sqlalchemy import text
query = text("""
SELECT
    product_id,
    product_title,
    COUNT(*) AS awesome_count
FROM
    product_reviews
WHERE
    review_body LIKE '%awesome%'
GROUP BY
    product_id,
    product_title
ORDER BY
    awesome_count DESC;
    """)

awful_products_df = pd.read_sql(query, engine)
display(awful_products_df)
```

```python
# Task: Find reviews with high total votes but low helpful votes percentage.
# Outcome: Identify reviews that are highly voted but not considered helpful by users.

from sqlalchemy import text
query = text("""
SELECT
    review_id,
    (SUM(helpful_votes) / SUM(total_votes)) * 100 AS helpful_percentage
FROM
    product_reviews

GROUP BY
    review_id
HAVING
    helpful_percentage < 50

ORDER BY
     helpful_percentage;
    """)

not_useful_reviews_df = pd.read_sql(query, engine)
display(not_useful_reviews_df)
```

```python
# Task: Identify the most reviewed products per year.
# Outcome: Identify product popularity trends over time.
from sqlalchemy import text
query = text("""
WITH yearly_counts AS (
    SELECT
        product_id,
        product_title,
        YEAR(review_date) AS review_year,
        COUNT(*) AS review_count
    FROM product_reviews
    GROUP BY product_id, product_title, YEAR(review_date)
)
SELECT
    product_id,
    product_title,
    review_year,
    review_count,
    RANK() OVER (PARTITION BY review_year ORDER BY review_count DESC) AS popularity_rank
FROM yearly_counts
ORDER BY review_year DESC, popularity_rank ASC;
    """)

# SELECT
#     employee_name,
#     salary,
#     RANK() OVER (ORDER BY salary DESC) as salary_rank
# FROM employees;

most_reviewed_products_df = pd.read_sql(query, engine)
display(most_reviewed_products_df)
```

```python
# Task: Identify the users who wrote the most reviews.
# Outcome: Recognize the most active reviewers in the dataset.
from sqlalchemy import text
query = text("""
SELECT
    customer_id,
    COUNT(*) AS review_count
FROM
    product_reviews
GROUP BY
    customer_id
ORDER BY
    review_count DESC;
    """)

most_active_users_df = pd.read_sql(query, engine)
display(most_active_users_df)
```
